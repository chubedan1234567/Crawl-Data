```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
# click vào xem thêm để lấy thêm link sản phẩm
for i in range(1, 10):
    url = 'https://hoanghamobile.com/dien-thoai-di-dong?&p=' + str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

# lấy tất cả các đường link sản phẩm
links = soup.find_all('a', class_='title')

list_link = []
for link in links:
    # thêm địa chỉ trang web vào đường link
    url = 'https://hoanghamobile.com' + link['href']
    list_link.append(url)
print(list_link)
```
