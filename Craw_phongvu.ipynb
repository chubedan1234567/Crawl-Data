```python
# import thư viện bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
links = []

# click vào trang tiếp theo
for i in range(1, 15):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    url = 'https://phongvu.vn/c/laptop?page=' + str(i)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    url1 = 'https://phongvu.vn/'
    for link in soup.find_all('a', class_='css-pxdb0j', href=True):
        links.append(url1 + link['href'])
print(links)
```

```python
links1 = []

# click vào trang tiếp theo
for i in range(15, 29):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    url = 'https://phongvu.vn/c/laptop?page=' + str(i)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    url1 = 'https://phongvu.vn/'
    for link in soup.find_all('a', class_='css-pxdb0j', href=True):
        links.append(url1 + link['href'])
print(links1)
```

```python
# truy cập vào từng link lấy thông tin
list_name = []
list_price = []
list_mo_ta = []
list_thuong_hieu = []
for link in links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    # lấy tên
    name = soup.find('h1', class_='css-4kh4rf')
    list_name.append(name.text)
    # # lấy giá
    price = soup.find('div', class_='att-product-detail-latest-price')
    list_price.append(price.text)
    # # lấy mô tả
    mo_ta = soup.find('div', class_='css-1nv5d5l').text
    list_mo_ta.append(mo_ta)
    # # lấy mã vạch
    thuong_hieu = soup.find('span', class_='css-n67qkj').text
    list_thuong_hieu.append(thuong_hieu)
    
```

```python
# dùng re xét trong list_name để loại chữ "Liên hệ đặt hàng"
list_name = [re.sub(r'Liên hệ đặt hàng', '', i) for i in list_name]
list_name
```

```python
list_price
```

```python
# dùng re xét trong list_price để loại chữ "đ"
list_price = [re.sub(r'₫', '', i) for i in list_price]
list_price
```

```python
# kiểm tra trong list_price có phần tử nào không có giá trị không
list_price = [i if i != '' else '0' for i in list_price]
list_price
```

```python
# dùng re xét trong list_mo_ta để mỗi khi gặp dấu '-' thì ngắt ra
list_mo_ta = [re.split(r'-', i) for i in list_mo_ta]
list_mo_ta
```

```python
# dùng re xét trong list_mo_ta để loại vị trí đầu tiên mỗi list
list_mo_ta = [i[1:] for i in list_mo_ta]
list_mo_ta
```

```python
# từ 1 nhiều list trong list_mo_ta tạo ra thành nhiều list chứa các thuộc tính theo thứu tự trong list_mo_ta
list_cpu = []
list_man_hinh = []
list_ram = []
list_do_hoa = []
list_dung_luong = []
list_os = []
list_pin = []
list_trong_luong = []

# lấy các thuộc tính trong list_mo_ta

for i in list_mo_ta:
    list_cpu.append(i[0])
    list_man_hinh.append(i[1])
    list_ram.append(i[2])
    list_do_hoa.append(i[3])
    list_dung_luong.append(i[4])
    list_os.append(i[5])
    list_pin.append(i[6])
    list_trong_luong.append(i[7])

    
```
